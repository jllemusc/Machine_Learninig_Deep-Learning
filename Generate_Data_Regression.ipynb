Generate Data for regression

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

Generate the data

```python
X,y = make_regression(n_samples=175, n_features = 2, bias=0.3, random_state= 9)
```

Plot the data

```python
plt.figure(figsize=(8,5))
plt.plot(X,y,'go',label='Feature')
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.title('make_regression output')
plt.grid()
plt.show()
```
