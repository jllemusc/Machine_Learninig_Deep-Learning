**Generate Data MAke_Moons**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

Generate Data

```python
X,y = make_moons(n_samples = 185, noise = 0.2,random_state=99)
```

Plot Data

```python
fig, axs = plt.subplots(1,5, figsize=(18,3))
for i, ax in zip(range(5),axs):
  X,y = make_moons(n_samples = 125, noise=i*0.03, random_state= 85)
  for target, color, marker in zip (range(2),['y','k'],['*','^']):
    ax.scatter(X[y==target,0],X[y==target,1], c= color,marker=marker, label= 'class '+ format(target))
    ax.set_xlabel('Feature 1', fontsize= 20 )
    if(i==0):
      ax.set_ylabel('Feature 2', fontsize =20)
    ax.set_title('Moons, noise: ' + format(i*0.03))
    ax.grid()
plt.show
```
